```python
# import libraries

import pandas as pd

from pyathena import connect
```

```python
# define a PyAthena connection to a S3 folder to store results of the query

conn = connect(s3_staging_dir = 's3://sau-athena-queries/temp-folder-for-athena-queries/', 
               region_name = 'us-west-2')
```

```python
# Run query on a table using Pandas connecting to Athena (via PyAthena API)

df_web_rfmo = pd.read_sql("SELECT * FROM sau.web_rfmo", conn)
df_web_rfmo
```
